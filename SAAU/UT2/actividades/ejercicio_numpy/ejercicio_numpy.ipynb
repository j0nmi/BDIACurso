### Ejercicio 13

Compara los tiempos de ejecución entre una lista aleatoria de 1.000.000 de datos con una matriz aleatorio del mismo tamaño con Numpy.

```python
import random
import time

# Tiempo para la Lista
strartLista = time.time()
lista = [random.randint(1, 1000) for i in range(1000000)]
endLista  = time.time()
totalLista  = endLista - strartLista
print("Tiempo de ejecución para la lista aleatoria: " + str(totalLista))

# Tiempo para la Matriz
strartMatriz = time.time()
matriz = np.random.randint(1, 1000, size=(1000000,))
endMatriz = time.time()
totalMatriz = endMatriz - strartMatriz
print("Tiempo de ejecución para la matriz aleatoria: " + str(totalMatriz))
```

### Ejercicio 14

Genera dos matrices con dimensiones 10x5 y muestra la suma, resta y multiplicación matricial.

```python
matriz1 = np.random.randint(1, 11, size=(10, 5))  # Números aleatorios entre 1 y 10
matriz2 = np.random.randint(1, 11, size=(10, 5))  # Números aleatorios entre 1 y 10

print("Matriz 1:")
print(matriz1)
print("Matriz 2:")
print(matriz2)

suma = matriz1 + matriz2
print("Suma Matricial:")
print(suma)

resta = matriz1 - matriz2
print("Resta Matricial:")
print(resta)

multi = matriz1, matriz2
print("Multiplicación Matricial:")
print(multi)
```

### Ejercicio 15

Crea una matriz aleatoria de tamaño 4 , muestra por pantalla el valor minimo y maximo.
